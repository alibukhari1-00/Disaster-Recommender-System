# Japan Earth Quack Data 

```python
import pandas as pd
file=pd.read_csv("japan_earthquakes_1900-2024.csv")
file.head()
```

# Removing unwanted features

```python
columns_drop=["nst","gap","dmin","rms","updated","horizontalError","magNst","status","locationSource","magSource","net","id","magType"]
earthquack_dataset_japan_final=file.drop(columns=columns_drop)
earthquack_dataset_japan_final.head()
```

```python
earthquack_dataset_japan_final["Id"]= range(1, len(earthquack_dataset_japan_final) + 1)
```
